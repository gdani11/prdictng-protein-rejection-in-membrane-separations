# Ridge regression

```python
from sklearn.linear_model import LassoCV
from sklearn.linear_model import RidgeCV
from sklearn.linear_model import LinearRegression, Ridge, Lasso
```

```python
# Defining a range of alpha values
alphas = np.logspace(-4, -0.5, 30)

# Initializing lists to store MSE, RMSE, and R^2 for each alpha for training and test sets
train_mse = []
test_mse = []
train_rmse = []
test_rmse = []
train_r2 = []
test_r2 = []

# Computing MSE, RMSE, and R^2 for each alpha for both training and test sets
for alpha in alphas:
    ridge = Ridge(alpha=alpha)
    ridge.fit(X_train_scaledR, y_trainR)

    # Predictions on the training set
    y_train_pred = ridge.predict(X_train_scaledR)
    train_mse.append(mean_squared_error(y_trainR, y_train_pred))
    train_rmse.append(np.sqrt(train_mse[-1]))
    train_r2.append(r2_score(y_trainR, y_train_pred))

    # Predictions on the test set
    y_test_pred = ridge.predict(X_test_scaledR)
    test_mse.append(mean_squared_error(y_testR, y_test_pred))
    test_rmse.append(np.sqrt(test_mse[-1]))
    test_r2.append(r2_score(y_testR, y_test_pred))

# Finding the optimal alpha (where test MSE is minimum)
optimal_alpha_index = np.argmin(test_mse)
optimal_alpha = alphas[optimal_alpha_index]
optimal_mse_test = test_mse[optimal_alpha_index]
optimal_rmse_test = test_rmse[optimal_alpha_index]
optimal_r2_test = test_r2[optimal_alpha_index]

# Plotting MSE vs alpha for training and test sets
plt.figure(figsize=(10, 6))
plt.plot(alphas, train_mse, label='Train MSE', marker='o')
plt.plot(alphas, test_mse, label='Test MSE', marker='o')
plt.axvline(optimal_alpha, color='red', linestyle='--', label=f'Optimal alpha: {optimal_alpha}')
plt.scatter(optimal_alpha, optimal_mse_test, color='red')  # Mark the point of the optimal alpha
plt.xlabel('Alpha')
plt.ylabel('MSE')
plt.title('Error vs Alpha for Ridge Regression')
plt.xscale('log')
plt.legend()
plt.gca().invert_xaxis()
plt.show()

# Printing the optimal alpha and corresponding test MSE, RMSE, and R^2
print(f"Optimal alpha for Ridge regression: {optimal_alpha}")
print(f"Optimal test MSE for Ridge regression: {optimal_mse_test}")
print(f"Optimal test RMSE for Ridge regression: {optimal_rmse_test}")
print(f"Optimal test R^2 for Ridge regression: {optimal_r2_test}")
```

```python
# Function to plot actual vs predicted
def plot_actual_vs_predicted(model, X_test, y_test, model_name, tol):
    # Predict using the model
    y_pred = model.predict(X_test)
    # Calculate the difference between actual and predicted values
    diff = y_test - y_pred.flatten()
    df = pd.DataFrame({
        'Actual': y_test,
        'Predicted': y_pred,
        'Difference': np.abs(diff)
    })

    # Filter DataFrame based on tolerance
    df_small = df[df['Difference'] <= tol]
    df_large = df[df['Difference'] > tol]

    # Plot the results
    plt.figure(figsize=(10, 8))
    plt.scatter(df_small['Actual'], df_small['Predicted'], color='red', label='Predicted')
    plt.scatter(df_large['Actual'], df_large['Predicted'], color='red')
    plt.xlabel('Actual')
    plt.ylabel('Predicted')
    plt.plot([df['Actual'].min(), df['Actual'].max()], [df['Actual'].min(), df['Actual'].max()], color='blue', linestyle='--', label='Ideal')
    plt.title(f'Actual vs Predicted ({model_name})')
    plt.legend()
    plt.show()

    # Print out the number of points within and outside the tolerance
    print(f"Number of points within tolerance ({model_name}):", len(df_small))
    print(f"Number of points outside tolerance ({model_name}):", len(df_large))

# Set tolerance level
tolerance = 0.1

# Plot for each model
#plot_actual_vs_predicted(best_lr, X_test_scaledR, y_testR, 'Linear Regression', tolerance)
#plot_actual_vs_predicted(best_rf, X_test_scaledR, y_testR, 'Random Forest', tolerance)
#plot_actual_vs_predicted(best_xgb, X_test_scaledR, y_testR, 'XGBoost', tolerance)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import torch

def plot_actual_vs_predicted2(model, X_test, y_test, model_name, tol):
    model.eval()  # Set the model to evaluation mode
    with torch.no_grad():  # Turn off gradients for validation, saves memory and computations
        y_pred = model(X_test).flatten()

    # Converting tensors to numpy arrays for comparison and visualization
    y_pred_np = y_pred.numpy()
    y_test_np = y_test.flatten().numpy()

    # Calculate the difference between actual and predicted values
    diff = np.abs(y_test_np - y_pred_np)

    # Create DataFrame for easy handling of data
    df = pd.DataFrame({
        'Actual': y_test_np,
        'Predicted': y_pred_np,
        'Difference': diff
    })

    # Filter DataFrame based on tolerance
    df_small = df[df['Difference'] <= tol]
    df_large = df[df['Difference'] > tol]

    # Plot the results
    plt.figure(figsize=(10, 8))
    plt.scatter(df_small['Actual'], df_small['Predicted'], color='red', label='Predicted')
    plt.scatter(df_large['Actual'], df_large['Predicted'], color='red')
    plt.plot([df['Actual'].min(), df['Actual'].max()], [df['Actual'].min(), df['Actual'].max()], color='blue', linestyle='--', label='Ideal')
    plt.title(f'Actual vs Predicted ({model_name})')
    plt.xlabel('Actual')
    plt.ylabel('Predicted')
    plt.legend()
    plt.show()

    # Print out the number of points within and outside the tolerance
    print(f"Number of points within tolerance ({model_name}):", len(df_small))
    print(f"Number of points outside tolerance ({model_name}):", len(df_large))
```

```python
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
# Identify numeric columns (excludes one-hot encoded columns)
numeric_cols = X_trainR.select_dtypes(include=['float64', 'int64']).columns

# Scale numeric columns only
X_train_scaledR = X_trainR.copy()
X_test_scaledR = X_testR.copy()
X_train_scaledR[numeric_cols] = scaler.fit_transform(X_trainR[numeric_cols])
X_test_scaledR[numeric_cols] = scaler.transform(X_testR[numeric_cols])

# Convert boolean columns to int type
X_train_scaledR = X_train_scaledR.astype(float)
X_test_scaledR = X_test_scaledR.astype(float)

# Convert DataFrames to PyTorch tensors
X_train_tensor = torch.tensor(X_train_scaledR.values, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test_scaledR.values, dtype=torch.float32)
y_train_tensor = torch.tensor(y_trainR.values, dtype=torch.float32).view(-1, 1)
y_test_tensor = torch.tensor(y_testR.values, dtype=torch.float32).view(-1, 1)

# Create TensorDatasets and DataLoaders
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Define the neural network
class ANN(nn.Module):
    def __init__(self):
        super(ANN, self).__init__()
        self.fc1 = nn.Linear(56, 64)  # Adjust the input dimensions if necessary
        self.fc2 = nn.Linear(64, 32)
        self.fc3 = nn.Linear(32, 16)
        self.fc4 = nn.Linear(16, 1)   # Output layer for regression
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.relu(self.fc3(x))
        return self.fc4(x)

# Initialize the model
model = ANN()

# Loss and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

def train_model(model, train_loader, test_loader, criterion, optimizer, epochs=100):
    train_losses, test_losses = [], []
    train_r2_scores, test_r2_scores = [], []

    for epoch in range(epochs):
        model.train()
        total_loss = 0
        all_targets = []
        all_predictions = []

        for data, target in train_loader:
            optimizer.zero_grad()
            output = model(data)
            loss = criterion(output, target)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()

            # Collect all targets and outputs for R2 scoring
            all_targets.extend(target.tolist())
            all_predictions.extend(output.detach().numpy().flatten())

        train_loss = total_loss / len(train_loader.dataset)
        train_losses.append(train_loss)
        train_r2 = r2_score(all_targets, all_predictions)
        train_r2_scores.append(train_r2)

        model.eval()
        total_loss = 0
        all_targets = []
        all_predictions = []
        with torch.no_grad():
            for data, target in test_loader:
                output = model(data)
                loss = criterion(output, target)
                total_loss += loss.item()

                all_targets.extend(target.tolist())
                all_predictions.extend(output.numpy().flatten())

        test_loss = total_loss / len(test_loader.dataset)
        test_losses.append(test_loss)
        test_r2 = r2_score(all_targets, all_predictions)
        test_r2_scores.append(test_r2)

        print(f"Epoch {epoch+1}: Train Loss = {train_loss:.4f}, Test Loss = {test_loss:.4f}, Train R² = {train_r2:.4f}, Test R² = {test_r2:.4f}")

    return train_losses, test_losses, train_r2_scores, test_r2_scores

# Train the network and get the results including R²
train_losses, test_losses, train_r2_scores, test_r2_scores = train_model(model, train_loader, test_loader, criterion, optimizer)

# Plotting losses and R² scores
fig, ax = plt.subplots(1, 2, figsize=(15, 5))

ax[0].plot(train_losses, label='Training Loss')
ax[0].plot(test_losses, label='Testing Loss')
ax[0].set_title('Loss per Epoch')
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('Loss')
ax[0].legend()

ax[1].plot(train_r2_scores, label='Training R²')
ax[1].plot(test_r2_scores, label='Testing R²')
ax[1].set_title('R² Score per Epoch')
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('R² Score')
ax[1].legend()

plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor 
def plot_and_print_metrics(model, param_grid, X_train, y_train, X_test, y_test, model_name):
    grid_search = GridSearchCV(model, param_grid, cv=3, scoring='neg_mean_squared_error')
    grid_search.fit(X_train, y_train)
    best_model = grid_search.best_estimator_

    # Predictions and metrics calculation
    y_train_pred = best_model.predict(X_train)
    y_test_pred = best_model.predict(X_test)
    train_mse = mean_squared_error(y_train, y_train_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)
    train_rmse = np.sqrt(train_mse)
    test_rmse = np.sqrt(test_mse)
    train_r2 = r2_score(y_train, y_train_pred)
    test_r2 = r2_score(y_test, y_test_pred)

    print(f"{model_name} Best Parameters:", grid_search.best_params_)


    # Plotting hyperparameters effect
    for param, values in param_grid.items():
        train_errors, test_errors = [], []
        for value in values:
            params = {param: value}
            best_model.set_params(**params)
            best_model.fit(X_train, y_train)

            train_pred = best_model.predict(X_train)
            test_pred = best_model.predict(X_test)
            train_errors.append(mean_squared_error(y_train, train_pred))
            test_errors.append(mean_squared_error(y_test, test_pred))

        plt.figure(figsize=(10, 6))
        plt.plot(values, train_errors, label='Train MSE', marker='o')
        plt.plot(values, test_errors, label='Test MSE', marker='o')
        plt.xlabel(param)
        plt.ylabel('Mean Squared Error')
        plt.title(f'{model_name} MSE vs {param}')
        plt.legend()
        plt.grid(True)
        plt.show()

# Gradient Boosting configuration
gb_param_grid = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'n_estimators': [100, 200, 300],
    'min_samples_leaf': [1, 2, 4]
}
gb = GradientBoostingRegressor(random_state=42)

# XGBoost configuration
xgb_param_grid = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'n_estimators': [100, 200, 300],
    'min_child_weight': [1, 2, 4]
}
xgb = XGBRegressor(random_state=42)


# Execute the plotting and metrics function for both models
plot_and_print_metrics(gb, gb_param_grid, X_train_scaledR, y_trainR, X_test_scaledR, y_testR, "Gradient Boosting")
plot_and_print_metrics(xgb, xgb_param_grid, X_train_scaledR, y_trainR, X_test_scaledR, y_testR, "XGBoost")
```
